```python
%matplotlib notebook
import numpy as np
import torch

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import matplotlib.cm as cm
```

```python
n_epoch = 200

history ={}
history['x1'] = np.zeros((n_epoch))
history['x2'] = np.zeros((n_epoch))
history['E'] = np.zeros((n_epoch))
eta = 0.01
x = torch.tensor([2.0,2.0],requires_grad=True)
A = torch.tensor([[1.0,0.0],[0.0,1.0]])

for epoch in range(n_epoch):
    
    history['x1'][epoch] = x[0].detach().numpy()
    history['x2'][epoch] = x[1].detach().numpy()
    quad = torch.einsum("i,ij,j",x,A,x)
    history['E'][epoch] = quad.detach().numpy()
#     print(quad)
    quad.backward() 
    with torch.no_grad():
        x = x - eta * x.grad
    x.requires_grad = True 
```

# 描画用

```python
x = np.arange(-2, 2, 0.05) # x点として[-2, 2]まで0.05刻みでサンプル
y = np.arange(-2, 2, 0.05)  # y点として[-2, 2]まで0.05刻みでサンプル
x, y = np.meshgrid(x, y)  # 上述のサンプリング点(x,y)を使ったメッシュ生成
z = (x**2 + y**2)  #(x^2+y^2))  を計算してzz座標へ格納する。

fig = plt.figure()
ax = Axes3D(fig)
    
def init():
    return fig,

def animate(i):
    plt.cla()
    ax.set_aspect('equal')
    ax.set_xlabel('X1')
    ax.set_ylabel('X2')
    ax.set_zlabel('E(X)')
#     ax.view_init(elev=30., azim=3.6*-i)
    ax.view_init(elev=0.,azim=3.6*35) # 初期値(2,2)のときJust
    ax.plot_wireframe(x, y, z, color='blue',linewidth=0.3, alpha=0.4)
    
#     ax.plot(history['x1'][:i],history['x2'][:i],history['E'][:i],color='orange',lw=2,alpha=0.8)
    for j in range(i):
#         ax.scatter(history['x1'][j],history['x2'][j],history['E'][j], color='red',alpha=j/i)
        ax.scatter(history['x1'][j],history['x2'][j],history['E'][j], color=cm.Wistia(j/i))
    ax.scatter(history['x1'][i],history['x2'][i],history['E'][i], color='k',marker='x')

    return fig,

# Animate
ani = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=100, blit=True,)   
ani.save('anim_roll.gif', writer="imagemagick")

# plt.show() # 絵の出力。
```
